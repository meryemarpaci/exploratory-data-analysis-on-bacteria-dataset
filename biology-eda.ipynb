```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
excel_file = '/kaggle/input/biyoloji-data/Chlorellavulgaris.xlsx'
xls = pd.ExcelFile(excel_file)
```

```python
print("Sayfa isimleri:")
for sheet_name in xls.sheet_names:
    print(sheet_name)
```

```python
df1 = pd.read_excel(excel_file, sheet_name='Chlorella O.D.')
df2 = pd.read_excel(excel_file, sheet_name='Chlorella Protein')
df3 = pd.read_excel(excel_file, sheet_name='Chlorella Lipid')
df4 = pd.read_excel(excel_file, sheet_name='Chlorella Astaksantin')
df5 = pd.read_excel(excel_file, sheet_name='Chlorella Caretenoıd')
df6 = pd.read_excel(excel_file, sheet_name='Chlorella Klorofil a')
df7 = pd.read_excel(excel_file, sheet_name='Chlorella klorofil b')

df1.head(20)
```

```python
df1.set_index('Chlorella O.D.', inplace=True)

plt.figure(figsize=(12, 8))

for ortam in df1.index:
    plt.plot(df1.columns, df1.loc[ortam], marker='o', label=ortam)

plt.title('Chlorella O.D. Çoğalma Miktarı Zaman Serisi')
plt.xlabel('Gün')
plt.ylabel('Çoğalma Miktarı')
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(12, 8))

for ortam in df1.columns:
    plt.plot(df1.index, df1[ortam], marker='o', label=str(ortam) + ' - Chlorella klorofil b')

for ortam in df2.columns:
    plt.plot(df2.index, df2[ortam], marker='x', label=str(ortam) + ' - Chlorella klorofil a')

plt.title('Chlorella Klorofil b ve Klorofil a Değerleri Zaman Serisi')
plt.xlabel('Gün')
plt.ylabel('Değer')
plt.xticks(df1.index)  
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6, df7], keys=['Chlorella O.D.','Chlorella Protein','Chlorella Lipid','Chlorella Astaksantin','Chlorella Caretenoıd','Chlorella Klorofil a','Chlorella klorofil b'])
df.describe().T
```

```python
df.info()
```

```python
for key in df.index.levels[0]:
    group_df = df.loc[key]
    plt.figure(figsize=(10, 6))
    for column in group_df.columns:
        plt.plot(group_df.index, group_df[column], marker='o', label=column)

    plt.title(f'{key} Alt Grubu Görselleştirilmesi')
    plt.xlabel('Index')  # İndex değerlerine göre görselleştirme yapıldığı için
    plt.ylabel('Değer')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(12, 8))


bar_width = 0.1
index = np.arange(len(df.columns))
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

for i, key in enumerate(df.index.levels[0]):
    group_df = df.loc[key]
    plt.bar(index + i * bar_width, group_df.mean(), bar_width, label=key, color=colors[i])

plt.xlabel('Günler')
plt.ylabel('Ortalama Değer')
plt.title('Tüm Alt Grupların Ortalama Değerlerinin Görselleştirilmesi')
plt.xticks(index + bar_width * 3.5, df.columns, rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 8))

index = np.arange(len(df.columns))
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

for i, key in enumerate(df.index.levels[0]):
    group_df = df.loc[key]
    plt.scatter(index, group_df.mean(), label=key, color=colors[i])

plt.xlabel('Günler')
plt.ylabel('Ortalama Değer')
plt.title('Tüm Alt Grupların Ortalama Değerlerinin Nokta Grafikleri ile Görselleştirilmesi')
plt.xticks(index, df.columns, rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 8))

index = np.arange(len(df.columns))
colors = ['b', 'g', 'r', 'c', 'm', 'y']

valid_groups = [group for group in df.index.levels[0] if group != 'Chlorella Lipid']

for i, key in enumerate(valid_groups):
    group_df = df.loc[key]
    plt.plot(index, group_df.mean(), marker='o', label=key, color=colors[i])

plt.xlabel('Günler')
plt.ylabel('Ortalama Değer')
plt.title('Chlorella Lipid Dışındaki Ortalama Değerlerinin Çizgi ve Nokta Grafikleri ile Görselleştirilmesi')
plt.xticks(index, df.columns, rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 8))

valid_groups = [group for group in df.index.levels[0] if group != 'Chlorella Lipid']

for key in valid_groups:
    group_df = df.loc[key]
    sns.kdeplot(data=group_df.values.flatten(), label=key)

plt.title('Chlorella Lipid Dışındaki Alt Grupların Değerlerinin Yoğunluk Grafiği')
plt.xlabel('Değerler')
plt.ylabel('Yoğunluk')
plt.legend()
plt.grid(True)
plt.show()
```
